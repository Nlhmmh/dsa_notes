# Big O Notation

Big O Notation is a mathematical notation used to describe the **upper bound of an algorithm's running time or space requirements** in terms of input data size (n). It provides a way to express the efficiency of an algorithm, allowing for comparison between different algorithms based on their performance.

Given 2 functions f(n) and g(n), we say that f(n) is O(g(n)) if there exist positive constants c and n0 such that for all n ≥ n0, the inequality f(n) ≤ c * g(n) holds true. This means that, beyond a certain point, the growth rate of g(n) will always be greater than or equal to the growth rate of f(n), up to a constant factor.

Rules for Big O Notation:
1. Ignore constant factors
2. Ignore lower-order terms

**Example 1**
```text
Given f(n) = 2n + 3
We got g(n) = n
To determine if f(n) is O(g(n)), we can find constants c and n0.
Let's choose c = 2 + 1 = 3.
f(n) <= c * g(n) for n ≥ 1
2n + 3 ≤ 3n
n0 ≥ 3
This means that beyond n0 = 3, f(n) will always be less than or equal to 3 times g(n).
Therefore, we can conclude that f(n) is O(g(n)).
```

````python
# Draw a graph to visualize the relationship between f(n) and g(n).

import matplotlib.pyplot as plt
import numpy as np

# Define the range of n
n = np.arange(0, 10, 1)

# Define the functions
c = 3
f_n = 2 * n + 3
g_n = c * n

# Create the plot
plt.plot(n, f_n, label='f(n) = 2n + 3', color='blue')
plt.plot(n, g_n, label='g(n) = c * n', color='orange')
plt.title('Big O Notation Visualization')
plt.xlabel('n')
plt.ylabel('Function Value')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
````

# Table of Common Big O Notations

| Type | Big O Notation | Description | Example Algorithms |
|------|----------------|-------------|-------------------|
| Constant | O(1) | The running time does not change with the size of the input data. | Hash table lookup |
| Logarithmic | O(log n) | The running time grows logarithmically with the input size. | Binary search |
| Linear | O(n) | The running time grows linearly with the input size. | Linear search, traversing a list |
| Linearithmic | O(n log n) | The running time grows in proportion to n log n. | Merge sort, quicksort |
| Quadratic | O(n^2) | The running time grows quadratically with the input size. | Bubble sort, insertion sort |
| Cubic | O(n^3) | The running time grows cubically with the input size. | Matrix multiplication |
| Exponential | O(2^n) | The running time doubles with each additional input element. | Recursive Fibonacci, brute-force algorithms |
| Factorial | O(n!) | The running time grows factorially with the input size. | Traveling salesman problem, permutations |



````python
# Visualizing Big O Notation

import matplotlib.pyplot as plt
import numpy as np
import math

# Define the range of n
n = np.arange(0, 30, 1)

# Define the functions
f_constant = np.full_like(n, 1)  # O(1)
f_logarithmic = np.log(n + 1)  # Adding 1 to avoid log(0)
f_linear = n
f_linearithmic = n * np.log(n + 1)  # Adding 1 to avoid log(0)
f_quadratic = n**2
f_cubic = n**3
f_exponential = 2**n
f_factorial = np.array([math.factorial(i) for i in n])

# Create the plot
plt.figure(figsize=(12, 8))
plt.plot(n, f_constant, label='O(1)', color='blue')
plt.plot(n, f_logarithmic, label='O(log n)', color='orange')
plt.plot(n, f_linear, label='O(n)', color='green')
plt.plot(n, f_linearithmic, label='O(n log n)', color='red')
plt.plot(n, f_quadratic, label='O(n^2)', color='purple')
plt.plot(n, f_exponential, label='O(2^n)', color='pink')
plt.plot(n, f_factorial, label='O(n!)', color='gray')

# Put the labels directly on the curves
ax = plt.gca()
curve_labels = [
    (f_constant, 'O(1)', 'blue', 25),
    (f_logarithmic, 'O(log n)', 'orange', 25),
    (f_linear, 'O(n)', 'green', 25),
    (f_linearithmic, 'O(n log n)', 'red', 15),
    (f_quadratic, 'O(n^2)', 'purple', 5),
    (f_exponential, 'O(2^n)', 'pink', 5),
    (f_factorial, 'O(n!)', 'gray', 4),
]

for values, label, color, label_x in curve_labels:
    ax.annotate(
        label,
        xy=(label_x, values[label_x]),
        xytext=(-20, 0),
        textcoords='offset points',
        color=color,
        fontsize=10,
        va='center',
        bbox=dict(facecolor='white', alpha=0.75, edgecolor='none'),
    )

plt.xlabel('n')
plt.ylabel('Function Value')
plt.title('Big O Notation Comparison')
plt.ylim(0, 50)  # Limit y-axis for better visibility
plt.tight_layout()
plt.legend()
plt.show()
````
